# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

---

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display a sample of y
print(y.head())
```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns=["credit_ranking"])

# Review the features DataFrame
print(X.head())
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_train_scaled = scaler.fit_transform(X_train)

# Fit the scaler to the features training dataset
X_test_scaled = scaler.transform(X_test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
input_features = X_train_scaled.shape[1]

# Review the number of features
print(f"Number of input features: {input_features}")
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1 = 8

# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2 = 5

# Define the number of neurons in the output layer
output_neurons = 1
```

```python
# Create the Sequential model instance
model = Sequential()

# Add the first hidden layer
model.add(Dense(units=hidden_nodes_layer1, activation='relu', input_dim=input_features))

# Add the second hidden layer
model.add(Dense(units=hidden_nodes_layer2, activation='relu'))

# Add the output layer to the model specifying the number of output neurons and activation function
model.add(Dense(units=output_neurons, activation='sigmoid'))
```

```python
# Display the Sequential model summary
model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Fit the model using 50 epochs and the training data
model.fit(X_train_scaled, y_train, epochs=50, verbose=1)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
loss, accuracy = model.evaluate(X_test_scaled, y_test, verbose=1)

# Display the model loss and accuracy results
print(f"Model Loss: {loss}, Model Accuracy: {accuracy}")
```

### Step 4: Save and export your model to a keras file, and name the file `student_loans.keras`.


```python
# Set the model's file path
model_file_path = "student_loan_model.keras"

# Export your model to a keras file
model.save(model_file_path)
```

---
## Predict Loan Repayment Success by Using your Neural Network Model

### Step 1: Reload your saved model.

```python
# Set the model's file path
model_file_path = "student_loan_model.keras"

# Import the load_model function
from tensorflow.keras.models import load_model

# Load the model to a new object
loaded_model = load_model(model_file_path)
```

### Step 2: Make predictions on the testing data and save the predictions to a DataFrame.

```python
# Make predictions with the test data
predictions = loaded_model.predict(X_test_scaled)

# Display a sample of the predictions
print(predictions[:5])
```

```python
# Save the predictions to a DataFrame and round the predictions to binary results
predictions_df = pd.DataFrame((predictions > 0.5).astype(int), columns=["Predicted"])
```

### Step 4: Display a classification report with the y test data and predictions

```python
# Print the classification report with the y test data and predictions
print(classification_report(y_test, predictions_df))
```

**1. Describe the data that you would need to collect to build a recommendation system to recommend student loan options for students. Explain why this data would be relevant and appropriate.**

To build a recommendation system for student loan options, the following types of data would be necessary:

*Demographic Information: Age, location, education level, and employment status. This helps determine eligibility for specific loans and predict the likelihood of repayment.
*Financial Data: Credit score, annual income, existing debt, and savings. This is crucial for assessing loan affordability and risk.
*Loan Preferences: Interest rate sensitivity, loan amount needed, repayment term preferences. This ensures personalized recommendations.
*Academic Information: School name, degree program, expected graduation date. Some loans are specific to certain schools or degrees.
*Behavioral Data: Previous loan history, payment patterns, and financial literacy level. This helps refine recommendations based on prior financial decisions.

**2. Based on the data you chose to use in this recommendation system, would your model be using collaborative filtering, content-based filtering, or context-based filtering? Justify why the data you selected would be suitable for your choice of filtering method.**

The recommendation system would primarily use content-based filtering.

Justification:

*Content-based filtering relies on the attributes of loans (interest rates, repayment terms, eligibility criteria) and the user's financial profile to make recommendations.
*Since loan recommendations are highly personalized based on creditworthiness and individual preferences, it is more effective to match users with loans based on their financial and demographic data rather than relying on other users' choices.
*Collaborative filtering (which recommends based on what similar users have chosen) would not be ideal because student loans are highly individualized, and one person’s loan suitability does not necessarily apply to another.
*Context-based filtering could also play a role, especially if additional factors like economic conditions, job market trends, and tuition costs are included, but content-based filtering would be the primary approach.

**3. Describe two real-world challenges that you would take into consideration while building a recommendation system for student loans. Explain why these challenges would be of concern for a student loan recommendation system.**

Challenge 1: Data Privacy and Security

*Financial and personal data are highly sensitive, and any data breach could have serious consequences for users.
*Strict compliance with regulations such as GDPR (General Data Protection Regulation) and CCPA (California Consumer Privacy Act) would be necessary.
*Secure storage and encryption of financial records would be essential to protect users from identity theft and fraud.

Challenge 2: Bias and Fairness in Loan Recommendations

*Machine learning models may develop bias based on historical lending patterns, which could unfairly disadvantage certain demographic groups.
*Ensuring that the model does not discriminate based on race, gender, or socioeconomic status is critical for ethical lending practices.
*Fairness metrics and regulatory oversight would be needed to prevent discriminatory lending practices and ensure equal access to loans.
